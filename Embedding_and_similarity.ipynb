```python
#TOKENIZING AND PERFORMING EMBEDDING FOR ALL Job DESCRIPTION FILES
import os
import json
from nltk.tokenize import word_tokenize
from sentence_transformers import SentenceTransformer

# Initialize the SentenceTransformer model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Define the folder path containing the job description JSON files
folder_path = "/job_description_records"

# Initialize a list to store the embeddings for each JSON file
job_embeddings = []

# Iterate through the JSON files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)

        with open(file_path, "r") as json_file:
            # Load the JSON data
            data = json.load(json_file)

        job_descriptions = data
        columns_to_tokenize = [
            "Job Description",
            "Core Responsibilities",
            "Required Skills",
            "Educational Requirements",
            "Preferred Qualifications",
            "Compensation and Benefits",
        ]
        final_job_desc = []

        # Tokenize text in specified columns
        for job_desc in job_descriptions:
            for column in columns_to_tokenize:
                column_text = job_desc.get(column, "")
                column_tokens = word_tokenize(column_text)
                final_job_desc.extend(column_tokens)

        # Join the tokens into a single string for this JSON file
        result_string = ' '.join(final_job_desc)

        # Encode the text and store the embedding
        job_embedding = model.encode(result_string)
        job_embeddings.append(job_embedding)

# Now, job_embeddings contains the embeddings for each JSON file
# You can access the embeddings for each file using job_embeddings[index]
# For example, job_embeddings[0] contains the embedding for the first JSON file.
len(job_embeddings)
```

```python
#TOKENIZING AND PERFORMING EMBEDDING ON ALL CV json FILES
import os
import json
import nltk
import numpy as np
from nltk.tokenize import word_tokenize
from sentence_transformers import SentenceTransformer

# Initialize the NLTK tokenizer
nltk.download("punkt")

# Define the folder path containing the CV JSON files
folder_path = "/parsed_resumes"

# Initialize a list to store the tokenized and preprocessed CV data
cv_data_list = []

# Initialize a list to store embeddings
cv_embeddings_list = []

# Initialize the SentenceTransformer model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

# Iterate through the JSON files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)

        with open(file_path, "r") as json_file:
            # Load the JSON data
            data = json.load(json_file)

        json_data = data

        # Tokenize the "skills" section
        skills_tokens = [word_tokenize(skill) for skill in json_data["skills"]]

        # Tokenize the "experience" section
        try:
            experience_tokens = [word_tokenize(exp) for exp in json_data["experience"]]
        except :
          continue


        # Tokenize the "category" section
        category_tokens = word_tokenize(json_data["category"])

        # Tokenize the "education" section
        education_tokens = [word_tokenize(education_item["Institution"]) for education_item in json_data["education"]]
        degree_tokens = [word_tokenize(education_item["Degree"]) for education_item in json_data["education"]]
        education_tokens.extend(degree_tokens)

        # Combine all tokenized sections into a single list
        cv_tokens = []
        cv_tokens.extend(skills_tokens)
        cv_tokens.extend(experience_tokens)
        cv_tokens.append(category_tokens)
        cv_tokens.extend(education_tokens)

        # Flatten the list and join into a single string
        cv_result_string = ' '.join([' '.join(desc) for desc in cv_tokens])

        # Append the preprocessed CV data to the list
        cv_data_list.append(cv_result_string)

        # Compute the embedding for the CV data
        cv_embedding = model.encode([cv_result_string])

        # Append the embedding to the list
        cv_embeddings_list.append(cv_embedding)

# Now, cv_data_list contains the preprocessed CV data for each JSON file in the folder.
# cv_embeddings_list contains the embeddings for the corresponding CV data.
len(cv_embeddings_list)
```

```python
import os
#FUNCTION to map json file names to array indexes
def list_json_files_in_folder(folder_path):

    # Initializing an empty list to store JSON file names
    json_file_names = []

    # Check if the folder path exists
    if os.path.exists(folder_path) and os.path.isdir(folder_path):
        # List files in the folder
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                # Check if the file has a .json extension
                if file.endswith(".json"):
                    # Append the JSON file name to the list
                    json_file_names.append(file)

    return json_file_names

# Example usage:
folder_path = "/content/gdrive/MyDrive/parsed_resumes"
json_file_names = list_json_files_in_folder(folder_path)
print(json_file_names)
```

```python
top_5_values_with_indices = []

# Iterate through each list in final_similarities
for index, similarity_list in enumerate(final_similarities):
    # Enumerate the list to keep track of original indexes
    enumerated_list = list(enumerate(similarity_list))

    # Sort the enumerated list by similarity values in descending order
    sorted_list = sorted(enumerated_list, key=lambda x: x[1], reverse=True)

    # Get the top 5 values with their original indexes
    top_5 = sorted_list[:5]

    # Append the result to the final list
    top_5_values_with_indices.append(top_5)

# Print the top 5 values with their original indexes for each list
print("Top 5 values for each job description:")
for index, top_5_values in enumerate(top_5_values_with_indices):
    print(f"Job_description {index + 1}:")
    for original_index, value in top_5_values:
      print(f"similarity value: {value}, cv_Name: {json_file_names[original_index]}")
```
